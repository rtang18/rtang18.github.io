```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ce = pd.read_csv('engineering.csv')
```

```python
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral5
from bokeh.plotting import figure
from bokeh.sampledata.autompg import autompg as df
from bokeh.transform import factor_cmap
```

```python
for i in range(len(ce)):
    major = ce.loc[i, "Major"]
    if major == "BioE/MSE Joint Major":
        ce.loc[i, "Major"] = "BioE/MSE"
    elif major == "Bioengineering":
        ce.loc[i, "Major"] = "BioE"
    elif major == "Civil Engineering":
        ce.loc[i, "Major"] = "CivE"
    elif major == "EECS/MSE Joint Major":
        ce.loc[i, "Major"] = "EECS/MSE"
    elif major == "EECS/NE Joint Major":
        ce.loc[i, "Major"] = "EECS/NE"
    elif major == "Electrical Eng & Comp Sci":
        ce.loc[i, "Major"] = "EECS"
    elif major == "Energy Engineering":
        ce.loc[i, "Major"] = "EnergyE"
    elif major == "Eng Math & Statistics":
        ce.loc[i, "Major"] = "EngMathStats"
    elif major == "Engineering Physics":
        ce.loc[i, "Major"] = "EngPhys"
    elif major == "Environmental Eng Science":
        ce.loc[i, "Major"] == "EnvEngSci"
    elif major == "Industrial Eng & Ops Rsch":
        ce.loc[i, "Major"] = "IEOR"
    elif major == "ME/NE Joint Major":
        ce.loc[i, "Major"] = "ME/NE"
    elif major == "MSE/ME Joint Major":
        ce.loc[i, "Major"] = "MSE/ME"
    elif major == "MSE/NE Joint Major":
        ce.loc[i, "Major"] = "MSE/NE"
```

```python
ce = ce[(ce['Major'] != "MSE/NE") & (ce['Major'] != "MSE/ME") & (ce['Major'] != 'ME/NE') & (ce['Major'] != 'EECS/MSE') & (ce['Major'] != 'EECS/NE') & (ce['Major'] != 'BioE/MSE')]
```

```python
gender_grouped = ce.groupby(by=["Academic Yr", "Major", "Gender Desc"]
                  ).sum().drop('Prorated Degrees Awarded Cnt', axis=1)
```

```python
year_majors = []
for i in gender_grouped.index:
    i = i[:2]
    year_majors.append(i)
year_majors
```

```python
gender_grouped = gender_grouped.reset_index()
```

```python
major = list(year_majors[0])
major + ['Decline to State', 0]
```

```python
j = 103
for i in year_majors:
    major = list(i)
    gender_grouped.loc[j] = major + ['Decline to State', 0]
    j += 1
    gender_grouped.loc[j] = major + ['Female', 0]
    j += 1
    gender_grouped.loc[j] = major + ['Male', 0]
    j += 1
```

```python
gender_grouped.tail()
```

```python
gender_grouped = gender_grouped.groupby(by=["Academic Yr", "Major", "Gender Desc"]).sum()
gender_grouped
```

```python
year_majors = []
for i in np.arange(0, len(gender_grouped), 3):
    major = gender_grouped.index[i][:2]
    year_majors.append(major)
```

```python
gender_grouped = gender_grouped.reset_index()
```

```python
from bokeh.core.properties import value
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
output_file("bar_stacked_grouped.html")

factors = year_majors

regions = ['dts', 'female', 'male']

source = ColumnDataSource(data=dict(
    x=factors,
    dts=gender_grouped.loc[gender_grouped['Gender Desc'] == 'Decline to State', 'Student Headcount'],
    female=gender_grouped.loc[gender_grouped['Gender Desc'] == 'Female', 'Student Headcount'],
    male=gender_grouped.loc[gender_grouped['Gender Desc'] == 'Male', 'Student Headcount']
))

p = figure(x_range=FactorRange(*factors), plot_height=750, plot_width = 1500,
           toolbar_location=None, tools="hover", tooltips="$name: @$name")

p.vbar_stack(regions, x='x', width=0.9, alpha=0.5, color=["purple","red", "blue"], source=source,
             legend=[value(x) for x in regions])

#p.y_range.start = 0
#p.y_range.end = 18
p.x_range.range_padding = 0.5
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "horizontal"

show(p)
```

```python
percentages = np.array([])
for i in range(0, len(gender_grouped), 3):
    total = gender_grouped.loc[i, "Student Headcount"] + gender_grouped.loc[i+1, "Student Headcount"] + gender_grouped.loc[i+2, "Student Headcount"]
    for k in range(0, 3):
        p = gender_grouped.loc[i+k, "Student Headcount"] / total
        percentages = np.append(percentages, p)
```

```python
gender_grouped["Percentage"] = percentages
```

```python
gender_grouped.head()
```

```python
from bokeh.core.properties import value
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
output_file("bar_stacked_grouped_percentages.html")

factors = year_majors

regions = ['dts', 'female', 'male']

source = ColumnDataSource(data=dict(
    x=factors,
    dts=gender_grouped.loc[gender_grouped['Gender Desc'] == 'Decline to State', 'Percentage'],
    female=gender_grouped.loc[gender_grouped['Gender Desc'] == 'Female', 'Percentage'],
    male=gender_grouped.loc[gender_grouped['Gender Desc'] == 'Male', 'Percentage']
))

p = figure(x_range=FactorRange(*factors), plot_height=500, plot_width = 1200,
           toolbar_location=None, tools="hover", tooltips="$name: @$name")

p.vbar_stack(regions, x='x', width=0.9, alpha=0.5, color=["purple","red", "blue"], source=source,
             legend=[value(x) for x in regions])

#p.y_range.start = 0
#p.y_range.end = 18
p.x_range.range_padding = 0.5
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "horizontal"

show(p)
```
